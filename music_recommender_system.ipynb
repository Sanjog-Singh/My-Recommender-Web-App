```python
import numpy as np
import pandas as pd
```

```python
musics_df=pd.read_csv('2420_musics.csv')
```

```python
musics_df.isnull().sum()
```

```python
musics_df.dropna(inplace=True)
```

```python
musics_df.duplicated().sum()
```

```python
musics_df=musics_df.drop_duplicates()
```

```python
musics_df.duplicated().sum()
```

```python
musics_df.shape
```

```python
musics_df.head()
```

```python
musics_df['User-Rating']
```

```python
l=[]
for i in musics_df['User-Rating']:
    l.append(i[:3])
l
```

```python
musics_df['User-Rating']=l
```

```python
musics_df['Album/Movie']=musics_df['Album/Movie'].str.replace(' ','')
musics_df['Singer/Artists']=musics_df['Singer/Artists'].str.replace(' ','')
musics_df
```

```python
musics_df['Singer/Artists']=musics_df['Singer/Artists'].str.replace(',',' ')
musics_df
```

```python
musics_df['tags']=musics_df['Singer/Artists']+musics_df['Genre']+musics_df['Album/Movie']+musics_df['User-Rating']
musics_df
```

```python
musics_df['tags'][0]
```

```python
final_df=musics_df[['Song-Name','tags']]
final_df
```

```python
final_df['tags']=final_df['tags'].apply(lambda x:x.lower())
```

```python
final_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=2391)
```

```python
vectors=cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
musics_similarity_scores=cosine_similarity(vectors)
```

```python
musics_similarity_scores[0]
```

```python
sorted(list(enumerate(musics_similarity_scores[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
final_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
  music_index=final_df[final_df['title']==music].index[0]
  distances=musics_similarity_scores[music_index]
  music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in music_list:
    print(final_df.iloc[i[0]].title)
```

```python
recommend('Aankh Marey')
```
