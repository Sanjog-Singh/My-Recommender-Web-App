## Popularity Based Recommender System

```python
# Merging Ratings with Book Details
ratings_with_name = ratings_df.merge(books_df,on='ISBN')
ratings_with_name.head()
```

```python
# Calculating Number of Ratings per Book
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Calculating Average Rating per Book
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
# Merging number of ratings and average rating dataframes into a single dataframe
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# Selecting books that have at least 250 ratings and are among the top 50 by average rating
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
# Merging to get book author and image URL; removing any duplicates
popular_df = popular_df.merge(books_df,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df.head()
```

## Collaborative Filtering Based Recommender System

```python
# Finding users who have rated more than 200 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
# Keeping ratings only from these active users
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
# Finding books that have received at least 50 ratings
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
# Filtering the dataset to include only these popular books
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
# Creating a matrix where rows are books, columns are users, and values are ratings
final_df = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final_df.fillna(0,inplace=True)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
books_similarity_scores = cosine_similarity(final_df)
```

```python
books_similarity_scores.shape
```

```python
# Function to recommend 5 similar books based on cosine similarity scores
def recommend_book(books):
    # index fetch
    book_index = np.where(final_df.index==books)[0][0]
    books_list = sorted(list(enumerate(books_similarity_scores[book_index])),key=lambda x:x[1],reverse=True)[1:6]
    
    recommended_book_names = []
    for i in books_list:
        item = []
        temp_df = books_df[books_df['Book-Title'] == final_df.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        recommended_book_names.append(item)
    
    return recommended_book_names
```

```python
recommend_book('1984')
```
